# **LIMPIEZA DE DATOS** #

**Se realizo una limpieza en **Acquisition_Cost** eliminando el simbolo de dolar y comas que pudieran hacer daño a la hora de hacer operaciones**

```python
data_csv['Acquisition_Cost'] = data_csv['Acquisition_Cost'].replace('$', '').replace(',', '').astype(float)
data_csv['Acquisition_Cost'].head()
```

**Se eliminaron duplicados que habian en campain_ID**

```python
data_csv= data_csv.drop_duplicates(subset='Campaign_ID', keep='first') 
data_csv.head()
```

Por ultimo decido guardar el archivo ya limpio y comprimirlo para que en vez de trabajar con 200 mil datos trabajar con 5000 datos asi al momento de hacer graficos se pudieran visualizar de una mejor manera

```python
data_csv_5000 = data_csv.head(5000)  
data_csv_5000.to_csv('marketing_5000.csv', index=False, encoding='utf-8')

data_csv.to_csv("marketing.csv", index=False)
print("¡Archivo limpiado y guardado con éxito!")
print(f"Archivo guardado con {len(data_csv_5000)} registros")
```
